# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]

plt.plot(x,y_1,label='Y values',
         color='red',linestyle='dashed')  # Spelling causes a long chunk of error. -.-
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

Use `plt.show()` to show plot.

Code uses American (U.S.) English rather than British (U.K.) English:
- Centre (U.K.) <-> center (U.S.).
- Colour (U.K.) <-> color (U.S.).

Alpha: transparency/ opacity from 0-1; for grid.

Legend location:
- Horizontal: left, right.
- Vertical: upper, lower.
- Others: center, best.

Note: Best refers to placing legend at most suitable location determined by Matplotlib.

### Things to note

#### 1. Abbreviations.

|Long form|Abbreviation|
|:-:|:-:|
|`color`|`c`|
|`linestyle`|`ls`|
|`linewidth`|`lw`|

```python
plt.plot(x,y_1,color='red',linestyle='dashed',linewidth=2)
```

```python
plt.plot(x,y_1,c='red',ls='dashed',lw=2)
```

#### 2. Interactive and non-interactive environments.

Jupyter is interactive environment, so will give output despite omitting `plt.show()`.\
Jupyter auto-prints last line.\
Good practice to include such that code works in non-interactive environments (e.g. command line).

#### 3. Plotting values have default values for styling parameters.

```python
plt.plot(x,y_1,y_2)
```

Styling is only for plot, still need add others to make it complete.

```python
plt.plot(x,y_1,y_2)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')  # Line exempted as no labels specified to be put in legend.

plt.show()
```

```python
plt.plot(x,y_1,y_2)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)

plt.show()
```

#### 4. Splitting arguments into separate lines to improve readability.

```python
plt.plot(x,y_1,color='red',linestyle='dashed',linewidth=2)
```

```python
plt.plot(x,y_1,label='Y values',
         color='red',linestyle='dashed')
```

#### 5. Order of specifying keyword arguments do not matter.

Keyword arguments: color, linewidth, etc..

## 1.2 Adding another plot

Add another `plot` to plot data of `y_2`:

```python
plt.plot(x,y_2,label='Y2 values',color='blue')
```

Combined plot of y_1 and y_2:

```python
x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]

plt.plot(x,y_1,label='Y1 values',color='red',linestyle='dashed')
plt.plot(x,y_2,label='Y2 values',color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```
