# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]

plt.plot(x,y_1,label='Y values',
         color='red',linestyle='dashed')  # Spelling causes a long chunk of error. -.-
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

Use `plt.show()` to show plot.

Code uses American (U.S.) English rather than British (U.K.) English:
- Centre (U.K.) <-> center (U.S.).
- Colour (U.K.) <-> color (U.S.).

Alpha: transparency/ opacity from 0-1; for grid.

Legend location:
- Horizontal: left, right.
- Vertical: upper, lower.
- Others: center, best.

Note: Best refers to placing legend at most suitable location determined by Matplotlib.

## 1.2 Adding another plot

Add another `plot` to plot data of `y_2`:

```python
plt.plot(x,y_2,label='Y2 values',color='blue')
```

Combined plot of y_1 and y_2:

```python
x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]

plt.plot(x,y_1,label='Y1 values',color='red',linestyle='dashed')
plt.plot(x,y_2,label='Y2 values',color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

## 1.3 Yet another plot but with error bars

Includes $x$ and $y$ error bars using `errorbar()`.

```python
y_3=[0,4,8,12,16,20,24]
x_error,y_error=.1,.75
plt.errorbar(x,y_3,xerr=x_error,yerr=y_error,label="Y3 with errors",color='green')
```

Combined plot with y_3:

```python
x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]
y_3=[0,4,8,12,16,20,24]
x_error,y_error=.1,.75

plt.plot(x,y_1,label='Y1 values',color='red',linestyle='dashed')
plt.plot(x,y_2,label='Y2 values',color='blue')
plt.errorbar(x,y_3,xerr=x_error,yerr=y_error,label="Y3 with errors",color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

Can include list of errors instead for each point.\
Constant errors used for all points in y_3.

# 2 Better with NumPy

Convert Python lists to NumPy arrays:

```python
import numpy as np

x=[0,1,2,3,4,5,6]
y_1=[0,2,4,6,8,10,12]
y_2=[0,3,6,9,12,15,18]

np_x=np.array(x)
np_y_1=np.array(y_1)
np_y_2=np.array(y_2)

plt.plot(np_x,np_y_1,color='red',linestyle='dashed',label='Y values')
plt.plot(np_x,np_y_2,color='blue',label='Y2 values')

plt.show()
```

## 2.1 Adding mathematical functions

Generates data-related mathematical functions easily.

Plot $x^2$ and $sin(x)$:

```python
x=np.array([0,1,2,3,4,5,6])

x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend()

plt.show()
```

Looks bad as $sin(x)$ lies between $\pm1$, but $x^2$ no bounds.\
Fix by adding another y-axis with the same x-axis.

## We need another axis!

Use `twinx()` to have multiple axes.

Normal plot:

```python
x=np.array([0,1,2,3,4,5,6])
x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')

plt.twinx()

plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend(loc='lower right')

plt.show()
```

### Things to note

2 `legend()` calls, 1 for each axis.\
Plot looks bad due to few points.\
Fix plot with `np.linespace()`.

`np.linspace()` plot:

```python
x=np.array([0,1,2,3,4,5,6])
x=np.linspace(0,6,100)  # Run after the variable to set linespacing for it.
x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')

plt.twinx()

plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend(loc='lower right')

plt.show()
```

# 3 Saving to disc

```python
x=np.linspace(0,6,100)
x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x,sin_x,label='sin x',color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png',dpi=150)
```

Use plot by saving to disk.\
Can be saved in many formats (e.g. PDF, JPEG, PNG) and at different resolutions.

Use `savefig()` to save plot.\
Specify format with extension of file name: `[filename].[extension]`.\
Specify resolution using dots-per-inch (dpi).

Code will save file in same directory/ folder.\
To save elsewhere, specify path details.

Save plot in Downloads:

```python
x=np.linspace(0,6,100)
x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x,sin_x,label='sin x',color='blue')
plt.legend(loc='lower right')

plt.savefig('C:\\Users\\User\\Downloads\simple-plot.png',dpi=150)
```

Very important to include code for the figure above the 'plt.savefig()' line.\
It will override the file with same name, so ensure file names are unique or backup made.\
Note: Single `\` to designate file name and `\\` to avoid escaping filename path.

Save plot in `.JPEG` format:

```python
x=np.linspace(0,6,100)
x2=x**2
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x,sin_x,label='sin x',color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.jpeg',dpi=150)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

Using data from Berkeley Earth website.\
Filename: `Complete_TAVG_daily.txt`.

```python
data=np.loadtxt('Complete_TAVG_daily.txt',skiprows=24)
date=data[:,0]
anomaly=data[:,-1]

plt.plot(date,anomaly,alpha=.5)
plt.ylim([-8,8])

plt.show()
```

Small alpha value used to soft plot colour.\
Plot range symmetrical in $y$ direction.

Add horizontal line at 0.\
`hlines()` require $y$-value with starting and ending values for $x$.

```python
data=np.loadtxt('Complete_TAVG_daily.txt',skiprows=24)
date=data[:,0]
anomaly=data[:,-1]

plt.plot(date,anomaly,alpha=.5)
plt.ylim([-8,8])
plt.hlines(0,date[0],date[-1],linestyle='--',color='grey')

plt.show()
```

## 4.2 Styles

Matplotlib offers standard style templates.\
E.g. of plot with 'fivethieryeight':

```python
plt.style.use('fivethirtyeight')

data=np.loadtxt('Complete_TAVG_daily.txt',skiprows=24)

date=data[:,0]
anomaly=data[:,-1]

plt.plot(date,anomaly,alpha=.5)
plt.hlines(0,date[0],date[-1],linestyle='--',color='grey')
plt.ylim([-8,8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to average from Jan 1951 to Dec 1980.)')

plt.show()
```
